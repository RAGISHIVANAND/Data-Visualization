```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv("/content/drive/MyDrive/google colab  files/stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"])

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"])

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"])
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors)
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.legend(labels=labels,loc="upper left")
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of Cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types(1982-2016)",fontsize=14,weight="bold")
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel","Petrol","Gas"]
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of Cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types(1982-2016)",fontsize=14,weight="bold")
plt.show()
```

```python
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel","Petrol","Gas"]
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of Cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types(1982-2016)",fontsize=14,weight="bold")
sns.despine()
plt.show()
```
