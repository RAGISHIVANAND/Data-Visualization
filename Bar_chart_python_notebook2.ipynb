```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("/content/drive/MyDrive/bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9, 6))
#plt.bar(x=df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rgbwymc")

plt.bar(x=df_used_cars["Brand"],
        height=df_used_cars["Cars Listings"],
        color=["#FF0000", "#00FF00",
               "#0000FF", "#FFFFFF",
               "#FFFF00", "#FF00FF",
               "#00FFFF"])

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
#plt.bar(x=df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rgbwymc")

plt.bar(x=df_used_cars["Brand"],
        height=df_used_cars["Cars Listings"],
        color=["midnightblue"])
plt.title("Cars Listings by Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
#plt.bar(x=df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rgbwymc")

plt.bar(x=df_used_cars["Brand"],
        height=df_used_cars["Cars Listings"],
        color=["midnightblue"])
plt.title("Cars Listings by Brand",fontsize = 16, fontweight ="bold")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
#plt.bar(x=df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rgbwymc")

plt.bar(x=df_used_cars["Brand"],
        height=df_used_cars["Cars Listings"],
        color=["midnightblue"])

plt.title("Cars Listings by Brand",fontsize = 16, fontweight ="bold")
plt.xticks(rotation=45,fontsize =13)
plt.yticks(fontsize=13)
plt.ylabel("Number of Listings",fontsize=13)
plt.show()
```
