```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_gspc_ftse_00_10 = pd.read_csv("/content/drive/MyDrive/google colab  files/line_chart_data.csv")
```

```python
df_gspc_ftse_00_10
```

```python
df_gspc_ftse_00_10["new_date"]=pd.to_datetime(df_gspc_ftse_00_10["Date"])
```

```python
df_gspc_ftse_00_10["new_date"]
```

```python
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["GSPC500"])
plt.show()
```

```python
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["FTSE100"])
plt.show()
```

```python
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["GSPC500"])
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["FTSE100"])
plt.show()
```

```python
labels=["GSPC500","FTSE100"]
plt.figure(figsize=(20,8))
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["GSPC500"])
plt.plot(df_gspc_ftse_00_10["new_date"],df_gspc_ftse_00_10["FTSE100"])
plt.title("GSPC500 VS FTSE100 RETURNS(2000-2010)",fontsize=14,fontweight="bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize="large")
plt.show()
```

SELECTING A PERIOD BETWEEN 01-07-2008 AND 31-12-2008

```python
df_gspc_ftse_00_10_H2_2008=df_gspc_ftse_00_10[(df_gspc_ftse_00_10.new_date>='2008-07-01')&
                                              (df_gspc_ftse_00_10.new_date<='2008-12-31')]
```

```python
df_gspc_ftse_00_10_H2_2008
```

```python
labels=["GSPC500","FTSE100"]
plt.figure(figsize=(20,8))
plt.plot(df_gspc_ftse_00_10_H2_2008["new_date"],df_gspc_ftse_00_10_H2_2008["GSPC500"],color="midnightblue")
plt.plot(df_gspc_ftse_00_10_H2_2008["new_date"],df_gspc_ftse_00_10_H2_2008["FTSE100"],color="crimson")
plt.title("GSPC500 VS FTSE100 RETURNS(H2 2008)",fontsize=14,fontweight="bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize="large")
plt.show()
```
